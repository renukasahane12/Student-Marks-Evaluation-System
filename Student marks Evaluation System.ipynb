```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

names = []
maths = []
science = []
english = []

n = int(input("Enter number of students: "))

for i in range(n):
    print("\nEnter details of student", i+1)

    name = input("Enter Name: ")
    m = int(input("Enter Maths Marks: "))
    s = int(input("Enter Science Marks: "))
    e = int(input("Enter English Marks: "))

    names.append(name)
    maths.append(m)
    science.append(s)
    english.append(e)


data = {
    "Name": names,
    "Maths": maths,
    "Science": science,
    "English": english
}

df = pd.DataFrame(data)


df["Total"] = df["Maths"] + df["Science"] + df["English"]
df["Percentage"] = df["Total"] / 3


def grade(p):
    if p >= 90:
        return "A"
    elif p >= 75:
        return "B"
    elif p >= 60:
        return "C"
    else:


        
        return "D"

df["Grade"] = df["Percentage"].apply(grade)


print("\nStudent Result Table")
print(df)


plt.figure(figsize=(6,4))
plt.bar(df["Name"], df["Percentage"])
plt.title("Student Percentage")
plt.xlabel("Students")
plt.ylabel("Percentage")
plt.show()


marks = df[["Maths", "Science", "English"]]
sns.heatmap(marks, annot=True, cmap="coolwarm")
plt.title("Subject Marks Heatmap")
plt.show()
```
